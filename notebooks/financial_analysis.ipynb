```python
def calculate_liquidity_solvency_ratios(bs_df):
    # Initialize results dataframe
    results = pd.DataFrame(index=bs_df.index)

    # Current Ratio = Current Assets / Current Liabilities
    results["Current Ratio"] = (
        bs_df["CURRENT ASSETS (Bn. VND)"] / bs_df["Current liabilities (Bn. VND)"]
    )

    # Quick Ratio = (Current Assets - Inventory) / Current Liabilities
    if "Net Inventories" in bs_df.columns and bs_df["Net Inventories"].sum() > 0:
        results["Quick Ratio"] = (
            bs_df["CURRENT ASSETS (Bn. VND)"] - bs_df["Net Inventories"]
        ) / bs_df["Current liabilities (Bn. VND)"]
    else:
        # Use "Inventories, Net" if available, otherwise assume 0 inventories
        inv_col = (
            "Inventories, Net (Bn. VND)" if "Inventories, Net (Bn. VND)" in bs_df.columns else 0
        )
        results["Quick Ratio"] = (bs_df["CURRENT ASSETS (Bn. VND)"] - bs_df[inv_col]) / bs_df[
            "Current liabilities (Bn. VND)"
        ]

    # Debt-to-Equity Ratio = Total Liabilities / Total Equity
    results["Debt-to-Equity Ratio"] = (
        bs_df["LIABILITIES (Bn. VND)"] / bs_df["OWNER'S EQUITY(Bn.VND)"]
    )

    return results
```
